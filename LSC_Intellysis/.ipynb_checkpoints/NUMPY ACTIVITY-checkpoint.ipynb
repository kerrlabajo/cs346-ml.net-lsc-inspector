#### Follow the steps to this simple NUMPY Activity

This Simple Activity will use Numpy as an easy Image processing tool

```python
#Fist Load the skimage library from scikit-image  
#if unknown do a pip install scikit-image
#this will load an image from a file and convert it to numpy data

#in this activity, you can choose any image you like to load....

from skimage import io
photo = io.imread('d:\\CITLOGO.jpg')
type(photo)

```

```python
#this will indicate image properties  (# of row pixels, # of col pixels, #data Depth  )
# 3 means R G B
photo.shape
```

```python
# we will then display the image using matplotlib
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
# lets now do some image processing! first by doing mirror image. Lets reverse the ROWs

plt.imshow(photo[::-1])  

# take note : means all elements in the data, this means :: says all roww
#and the  -1 means reverse step
# syntax  (start:stop:step)



```

```python
# what will this show us?
plt.imshow(photo[:, ::-1])
```

```python
#the result above shows a reverse in the columns
```

```python
# we can also get a subsection of the image by doing this
plt.imshow(photo[50:100,150:200])
```

```python
#we can compress the data image by 1/2 by doing this
plt.imshow(photo[::2, ::2])
```

```python
#take note of the value ranges
```

```python
# we can even do Trigonometrical process on image now easily
import numpy as np
photo
photo_sin=np.sin(photo)
print(photo_sin)
print(np.sum(photo))
print(np.mean(photo))
print(np.std(photo))
#many more 

```

```python
# we can even do color binary operations
photo_binary=np.where(photo>128, 255, 0)  
# think of this as a ternary operation if photo data is >100 then replace it with 255 else 0

plt.imshow(photo_binary)
```

```python
#lastly it can even do pixel movement
plt.imshow(photo[:,:,0].T)
#.T means transpose
```

```python
#Fist Load the skimage library from scikit-image  
#this will load an image from a file and convert it to numpy data

import numpy as np
from skimage import io
import matplotlib.pyplot as plt

photo = io.imread('citlogo.jpg')
type(photo)

#this will indicate image properties  (# of row pixels, # of col pixels, #data Depth  )
# 3 means R G B
photo.shape

# Convert the image to grayscale using NumPy
gray_photo = np.dot(photo[...,:3], [0.2989, 0.5870, 0.1140])

# Display the grayscale image using matplotlib
plt.imshow(gray_photo, cmap='gray')

plt.show()
```
